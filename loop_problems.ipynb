```python
#first 10 even numbers
count=0;
i=1;
while count<10:
  if(i%2==0):
    count+=1
    print(i)
  i+=1
```

```python
#even numbers between 1 and 10
i=1;
while i<=10:
  if(i%2==0):
    print(i)
  i+=1
```

```python
#prime number in range 1-10
i=1;
list7=[]
while i<11:
  list7=[]
  for j in range(1,i+1):
    if(i%j==0):
      list7.append(j)
  if(len(list7)==2):
    print(i)
  i+=1
```

```python
#first 10 prime numbers
i=1
count=0
list7=[]
while count<10:
  list7=[]
  for j in range(1,i+1):
    if(i%j==0):
      list7.append(j)
  if(len(list7)==2):
    print(i)
    count+=1
  i+=1
```

```python
#armstrong numbers from 1 to 500
k=1
while k<=500:
  m=str(k)
  list8=[]
  for i in m:
    list8.append(int(i)**3)
  res=sum(list8)
  if(res==k):
    print(k)
  k+=1
```

```python
#first five armstrong numbers
k=1
count=0
while count<5:
  m=str(k)
  list8=[]
  for i in m:
    list8.append(int(i)**3)
  res=sum(list8)
  if(res==k):
    print(k)
    count+=1
  k+=1
```
