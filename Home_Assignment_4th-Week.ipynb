1. Take input of age of 3 people by user and determine oldest and youngest among them

```python
a=int(input())
b=int(input())
c=int(input())
if(a>b and a>c):
  print("A is oldest and age is :",a)
elif(b>a and b>c):
  print("B is oldest and age is :",b)
else:
  print("C is oldest and age is :",c)

if(a<b and a<c):
  print("A is Yougest and age is :",a)
elif(b<a and b<c):
  print("B is Youngest and age is :",b)
else:
  print("C is Youngest and age is :",c)

```

2. A student will not be allowed to sit in exam if his/her attendence is less than 75%. Take following input from user
  A)Number of classes held
  B)Number of classes attended.
  And print
  percentage of class attended
  Is student is allowed to sit in exam or not.

```python
#number of classes conducted
total=int(input())
#number of classes attended
attend=int(input())
print("Attendance of student is :",(attend/total*100))
if((attend/total*100)<75):
  print("Student is not allowed to sit for exam")
else:
  print("Student is allowed to sit for exam")

```

3.print whether The given number is of one digited or two digited or three digited or more than three digited. TAKE THE INPUT FROM USER

```python
inp=int(input())
inp=inp//10
if(inp==0):
  print("number is single digited")
else:
  inp=inp//10
  if(inp==0):
    print("number is two digited")
  else:
    inp=inp//10
    if(inp==0):
      print("Number is three digited")
    else:
      print("number is more than three digited")
```
