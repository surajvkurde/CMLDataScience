```python

# Q1. 
# Check if a given key already exists in a dictionary
# input
# dictionaryName = {1: 10, 2: 20, 3: 30, 4: 40, 5: 50, 6: 60}
# make a function is_key_present that will check if key already exists
# is_key_present(5)
# is_key_present(9)

# output
# Key is present in the dictionary                                                                              
# Key is not present in the dictionary
dict1={1: 10, 2: 20, 3: 30, 4: 40, 5: 50, 6: 60}

def is_key_present(key1):
  pres=False
  for x in dict1.keys():
    if(x==key1):
      pres=True
      
  if pres:
       print("Key is present in the dictionary")
  else:
      print("Key is not present in the dictionary")

is_key_present(5)
is_key_present(9)


```

```python
#Q4. 
# Sum all the items in a dictionary
# Input {'data1':100,'data2':-54,'data3':247}
# Output 293

dict3={'data1':100,'data2':-54,'data3':247}
sum=0;
for i in dict3.keys():
  sum+=dict3[i]
print(sum)
```
