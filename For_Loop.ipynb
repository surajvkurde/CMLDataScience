# 4. Program to print the difference of given list

```python
list1=[1,3,5,7,9,11]
d=[]
diff=0
l=len(list1)
for i in range(0,l):
    if(i==0):
        diff=list1[i]
    else:
        diff=diff-list1[i]
        d.append(diff)
print(diff)
print(d)
```
