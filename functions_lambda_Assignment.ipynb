1. create a lambda function that multiplies argument x with argument y and print the result..

```python
mul=lambda x,y:x*y
print(mul(int(input()),int(input())))
```

2. Write a Python program to create a lambda function that adds 15 to a given number passed in as an argument.

```python
add15=lambda x:x+15
print(add15(int(input())))
```

3. create a lambda function that subtracts argument x with argument y and print the result and datatype

```python
mul=lambda x,y:x-y
rs=mul(int(input()),int(input()))
print(rs,type(rs))
```

4. Write a Python program to find whether a given string starts with a given character using Lambda.

```python
startStr=lambda x:True if x.startswith('c') else False
print(startStr("CML"))
```

5.Write a Python program to find intersection of two given arrays using Lambda.

```python
import numpy as np
arr1=[1,2,3,4]
arr2=[4,5,6,7,8]
intersect=list(filter(lambda x:x in arr1,arr2))
print(intersect)
```
